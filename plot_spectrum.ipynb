```python
%matplotlib inline
import astropy.io.ascii as asc
import numpy as np 
import matplotlib.pyplot as plt
```

### Plotting spectra

This notebook tells you how to plot spectra from the simulation grid for the Matthews et al. 2020 paper "Stratified disc wind models for the AGN broad-line region: ultraviolet, optical and X-ray properties". First let's read in the data using astropy. 

```python
# read in run number 52 which is model A from the paper using astropy
fname = "spectra/run52_thmin70_thmax85_rv1e19_f0p01_r1.spec"
spec_table = asc.read(fname)
```

```python
# make a plot of the spectrum
plt.plot(spec_table["Lambda"], spec_table["A20P0.50"])
plt.xlabel("Wavelength")
plt.ylabel("Flux at 10 Gpc")
```

We can also plot all the viewing angles on one plot

```python
for col in spec_table.colnames:
    if col[0] == "A":
        plt.plot(spec_table["Lambda"], spec_table[col])
        
plt.semilogy() # log axes so we can see dynamic range ok.
```

Finally, let's take a closer look at one of the spectra with BALs (75 degree viewing angle looking into the wind cone)

```python
plt.plot(spec_table["Lambda"], spec_table["A75P0.50"])
plt.xlim(1000,3000)    
plt.legend()
```
